```python
def transform(x):
    if 'GB' in x:
        return x[:len(x) - 2]
    elif 'TB' in x:
        y = str(int(float(x[:len(x) - 2]) * 1024))
        return y
    else:
        return x
```

```python
print(transform('2TB'))
```

```python
print(transform('256GB'))
```

```python
print(transform('abc'))
```

```python
def apply_transform(x):
    out = ''
    for s in x.split():
        out = out + transform(s) + ' '
    return out.rstrip()
```

```python
apply_transform("256GB SSD + 2TB HDD")
```

```python
apply_transform("256GB SSD + 1.0TB HDD")
```

```python
new_df['Memory'] = new_df['Memory'].apply(apply_transform)
```

```python
new_df.sample(10)
```

```python
def calculate_SSD(x):
    ans = 0
    prev = ''
    for s in x.split():
        if s == 'SSD':
            return int(prev)
        prev = s
    return ans
```

```python
calculate_SSD('256 SSD + 1024 HDD')
```

```python
def calculate_HDD(x):
    ans = 0
    prev = ''
    for s in x.split():
        if s == 'HDD':
            return int(prev)
        prev = s
    return ans
```

```python
def calculate_flash(x):
    ans = 0
    prev = ''
    for s in x.split():
        if s == 'Flash':
            return int(prev)
        prev = s
    return ans
```

```python
def calculate_hybrid(x):
    ans = 0
    prev = ''
    for s in x.split():
        if s == 'Hybrid':
            return int(prev)
        prev = s
    return ans
```

```python
new_df['SSD'] = new_df['Memory'].apply(calculate_SSD)
new_df['HDD'] = new_df['Memory'].apply(calculate_HDD)
new_df['Flash Storage'] = new_df['Memory'].apply(calculate_flash)
new_df['Hybrid'] = new_df['Memory'].apply(calculate_hybrid)
```
